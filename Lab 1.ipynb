```python
def NN1(x, c, d):
    x1 = x[0]; x2 = x[1]; c1 = c[0]; c2 = c[1]; d1 = d[0]; d2 = d[1]
    return np.exp(-(c1/(1 + d1*(x1*x2)**2))-(c2/(1 + d1*(x1/x2)**2)))
```

```python
def NN2(x, b, d):
    x1 = x[0]; x2 = x[1]; b1 = b[0]; b2 = b[1]; d1 = d[0]; d2 = d[1]
    return np.exp(-(d1/(1 + b1*x1**2)*(1 + b2*x2**2))-((d2*(1+b1*x1**2))/(1+b2*x2**2)))
```

```python
def NN3(x, b, d):
    x1 = x[0]; x2 = x[1]; b1 = b[0]; b2 = b[1]; d1 = d[0]; d2 = d[1]
    value1 = 1/(1 + b1 * np.exp(-(d1*(x1*x2)**2) -(d2*(x1/x2)**2)))
    value2 = 1/(1 + b1 * np.exp(-(d1*(x1*x2)**2) -(d2*(x1/x2)**2)))
    return [value1, value2]
```

```python
# Redifing the neural network with the stated parameters

def NN1(x, c = [0.5, 0.1], d = [0.2, 0.6]):
    x1 = x[0]; x2 = x[1]; c1 = c[0]; c2 = c[1]; d1 = d[0]; d2 = d[1]
    return np.exp(-(c1/(1 + d1*(x1*x2)**2))-(c2/(1 + d2*(x1/x2)**2)))

def NN2(x, b = [0.1, 0.9], d = [0.9, 0.7]):
    x1 = x[0]; x2 = x[1]; b1 = b[0]; b2 = b[1]; d1 = d[0]; d2 = d[1]
    return np.exp(-(d1/(1 + b1*x1**2)*(1 + b2*x2**2))-((d2*(1+b1*x1**2))/(1+b2*x2**2)))

def NN3(x, b = [0.9, 0.6], d = [0.5, 0.9]):
    x1 = x[0]; x2 = x[1]; b1 = b[0]; b2 = b[1]; d1 = d[0]; d2 = d[1]
    value1 = 1/(1 + b1 * exp(-(d1*(x1*x2)**2) -(d2*(x1/x2)**2)))
    value2 = 1/(1 + b1 * exp(-(d1*(x1*x2)**2) -(d2*(x1/x2)**2)))
    return [value1, value2]
```

```python
def NN1(x, p = [0.5, 0.1, 0.2, 0.6]):
    c = [p[0], p[1]]; d = [p[2], p[3]]
    x1 = x[0]; x2 = x[1]; c1 = c[0]; c2 = c[1]; d1 = d[0]; d2 = d[1]
    return np.exp(-(c1/(1 + d1*(x1*x2)**2))-(c2/(1 + d2*(x1/x2)**2)))
```

*Comment*: Next I'll define the mse funtion. I'm using the noised output vector as the "true" outcome y. I then want to check if the network, for a given set of parameters, can correctly guess a similar output y_hat given the same input x. 

```python
# MSE function
def MSE_NN1(p):
    y = sim_1
    y_hat = NN1(x, p)
    e = y - y_hat
    mse = (np.matmul(np.transpose(e), e))/len(e)
    return mse
```

```python
p = [0.5, 0.1, 0.2, 0.6]
MSE_NN1(p)
```

*Comment*: When using the same parameters that we used to generate our data, mse becomes very low. Let's we happens if we give it different parameters.

```python
p = [5, 4, 3, 2]
MSE_NN1(p)
```

*Comment*: As expected, MSE becomes higher.
