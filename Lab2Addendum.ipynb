As we will see above one has to be careful with the numerical result evaluating eigenvalues of the positive definite matrices in high-dimensions. 

Computing the Gaussian likelihood in high dimensions has to be approached with care. Using the densities is dangerous since their values can be so close to zero that the logarithm function may crash. Also the singularity of the covariance has to be taken care. Below `llGaussian` is doing a better job than the previously used in low dimensions `llGaussian1` (see Lab 0). 

```python
#Evaluation of the inverse and the eigenvalues
def preceigval(Sigma):
    sigma2=np.diag(Sigma)
    w, v = np.linalg.eig(Sigma)
    if min(sigma2)==0 or min(w)<=0: #to protect against singularities
        delta=min(sigma2[sigma2>0])
        Sigma=Sigma+delta*np.identity(d)
    w, v = np.linalg.eig(Sigma)
    precision = np.linalg.inv(Sigma)
    return precision, w

precision_tshirt, w_tshirt = preceigval(cov_tshirt) 
precision_trouser, w_trouser = preceigval(cov_trouser) 
precision_pullover, w_pullover = preceigval(cov_pullover) 
precision_dress, w_dress = preceigval(cov_dress) 
precision_sneaker, w_sneaker = preceigval(cov_sneaker) 
precision_sandal, w_sandal = preceigval(cov_sandal) 
precision_coat, w_coat = preceigval(cov_coat) 
precision_shirt, w_shirt = preceigval(cov_shirt) 
precision_bag, w_bag = preceigval(cov_bag) 
precision_boot, w_boot = preceigval(cov_boot) 
```

```python




from scipy.stats import multivariate_normal

def llGaussian1(x,mu,Sigma): 
    ll=np.log(multivariate_normal.pdf(x, mean=mu, cov=Sigma)) 
    return ll

#Evaluation of the inverse and the eigenvalues
def preceigval(Sigma):
    sigma2=np.diag(Sigma)
    w, v = np.linalg.eig(Sigma)
    if min(sigma2)==0 or min(w)<=0: #to protect against singularities
        delta=min(sigma2[sigma2>0])
        Sigma=Sigma+delta*np.identity(d)
    w, v = np.linalg.eig(Sigma)
    precision = np.linalg.inv(Sigma)
    return precision, w

def llGaussian(x,mu,precision,w):
    d=len(mu)
    ll=-d*np.log(2*np.pi)/2 
    ll=ll-sum(np.log(np.real(w)))/2
    ll=ll-np.dot(np.dot(x-mu,precision),x-mu)/2
    return ll

#Testing the new way of evaluating the log-likelihood against the old way for low dimensions. 
mu = [2, -3]
Sigma = [[1, 0], [0, 10]]

barX=np.mean(X, axis=0)

X= np.random.multivariate_normal(mu, Sigma, 500)

barX=np.mean(X, axis=0)
S=np.cov(np.transpose(X))

prec, w = preceigval(S)

print(llGaussian(X[0],barX,prec,w))
print(llGaussian1(X[0],barX,S))
```

```python
def classmle(x):
    ll=np.zeros(10)
    ll[0]=llGaussian(x, mean_tshirt, precision_tshirt, w_tshirt)
    ll[1]=llGaussian(x, mean_trouser, precision_trouser, w_trouser)
    ll[2]=llGaussian(x, mean_pullover, precision_pullover, w_pullover)
    ll[3]=llGaussian(x, mean_dress, precision_dress, w_dress)
    ll[4]=llGaussian(x, mean_coat, precision_coat, w_coat)
    ll[5]=llGaussian(x, mean_sandal, precision_sandal, w_sandal)
    ll[6]=llGaussian(x, mean_shirt, precision_shirt, w_shirt)
    ll[7]=llGaussian(x, mean_sneaker, precision_sneaker, w_sneaker)
    ll[8]=llGaussian(x, mean_bag, precision_bag, w_bag)
    ll[9]=llGaussian(x, mean_boot, precision_boot, w_boot)
    return np.argmax(ll)
    
classmle(mean_sandal)
```
