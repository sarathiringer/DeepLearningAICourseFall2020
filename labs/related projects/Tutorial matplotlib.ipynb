## Pyplot tutorial
https://matplotlib.org/tutorials/introductory/pyplot.html

```python
import matplotlib.pyplot as plt
```

```python
# Generating a plot where matplotlib will assume sequence of x-axis values
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

```python
# Provide function with x-coordinates as well
plt.plot([1, 2, 3, 4], [1, 4,9, 16]) # First vector is x
plt.show()
```

```python
# Formatting the style
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') # 'ro' indicated reds and dots
plt.axis([0, 6, 0, 20]) # xmin, smax, ymin, ymax
plt.show()
```

```python
# Working with arrays instead
import numpy as np

# Sample evenly
t = np.arange(0., 5., 0.2)

# Red dashes, blur squares and green triangles
plt.plot(t, t, 'r--', t, np.cos(t), 'bs', t, np.sin(t), 'g^')
plt.show()
```

```python
# Plotting with keyword srings
data = {'a': np.arange(50),
       'c': np.random.randint(0, 50, 50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['a']) * 8

# Plotting a to x axis, b to y axis, c to colour and d to size
plt.scatter('a', 'b', c = 'c', s = 'd', data = data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```
